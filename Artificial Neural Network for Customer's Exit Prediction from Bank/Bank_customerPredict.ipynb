```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv("C:/Users/ARCHANA/Desktop/Data Scientist/Deep Learning/Bank customer churn prection ANN/BankCustomers.csv")
```

```python
import pandas as pd
import numpy as np

# Assuming 'dataset' is already loaded as a Pandas DataFrame
# Extract independent (X) and dependent (y) variables
X = dataset.iloc[:, 3:13]  # Extract features as a DataFrame
y = dataset.iloc[:, 13].values  # Extract target as a NumPy array

# Check if required columns exist in the DataFrame
if 'Geography' in X.columns and 'Gender' in X.columns:
    # Convert categorical features into dummy variables
    states = pd.get_dummies(X['Geography'], drop_first=True)
    gender = pd.get_dummies(X['Gender'], drop_first=True)

    # Concatenate the dummy variables with the original DataFrame
    X = pd.concat([X, states, gender], axis=1)

    # Drop the original categorical columns
    X = X.drop(['Geography', 'Gender'], axis=1)
else:
    print("Error: Columns 'Geography' or 'Gender' are missing from the DataFrame.")

# Ensure X is in the correct format (convert to NumPy array if needed for ML models)
X = X.values

```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Importing the required libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import confusion_matrix, accuracy_score

# Initializing the Artificial Neural Network (ANN)
classifier = Sequential()

# Adding the input layer and the first hidden layer
classifier.add(
    Dense(
        activation="relu", 
        input_dim=11,  # Number of input features
        units=6,  # Number of neurons in the first hidden layer
        kernel_initializer="uniform"  # Initializer for weights
    )
)

# Adding the second hidden layer
classifier.add(
    Dense(
        activation="relu", 
        units=6,  # Number of neurons in the second hidden layer
        kernel_initializer="uniform"
    )
)

# Adding the output layer
classifier.add(
    Dense(
        activation="sigmoid",  # Sigmoid for binary classification
        units=1,  # Single neuron for binary output
        kernel_initializer="uniform"
    )
)

# Compiling the ANN
classifier.compile(
    optimizer="adam",  # Optimizer for gradient descent
    loss="binary_crossentropy",  # Loss function for binary classification
    metrics=["accuracy"]  # Track accuracy during training
)

# Fitting the ANN to the Training set
classifier.fit(
    X_train,  # Features of the training set
    y_train,  # Target variable of the training set
    batch_size=10,  # Number of samples per gradient update
    epochs=100  # Number of complete passes through the training dataset
)

# Making the predictions on the Test set
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)  # Convert probabilities to binary predictions

# Generating the Confusion Matrix and calculating accuracy
cm = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)

# Output the results
print("Confusion Matrix:")
print(cm)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
